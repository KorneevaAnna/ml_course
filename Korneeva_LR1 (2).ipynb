## 1. Текстовое описание набора данных

Датасет "120 years of Olympic history: athletes and results" (https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results) содержит информацию об участниках Олимпийских игр с 1896 г. по 2016 г.  
  
###### Список полей
* ID: уникальный номер участника Олимпийских игр (тип: целое число).
* Name: имя участника Олимпийский игр (тип: строковый).
* Sex: пол участника Олимпийсих игр (тип: строковый).
* Age: возраст участника Олимпийсих игр (тип: вещественное число).
* Height: рост участника Олимпийсих игр (тип: вещественное число).
* Weight: вес участника Олимпийсих игр (тип: вещественное число).
* Team: наименование сборной, от имени которой выступает участник Олимпийсих игр (тип: строковый).
* NOC: код Национального олимпийского комитета (тип: строковый).
* Games: год и время года, соответствующие датам проведения Олимпийских игр (тип: строковый).
* Year: год проведения Олимпийсих игр (тип: целое  число).
* Season: время года, соответствующие датам проведения Олимпийских игр (тип: строковый).
* City: город проведения Олимпийских игр (тип: строковый).
* Sport: вид спорта, в котором выступает спортсмен (тип: строковый).
* Event: наименование соревнования (тип: строковый).
* Medal: тип медали, которую в указанном соревновании завоевал спортсмен (тип: строковый).

## 2. Предобработка данных

Разобъём предобработку данных на логические блоки.  
1. Удалим поле Name как неинформативное, т.к. каждый участник Олимпийских игр однозначно описывается своим ID. Также удалим поле Games, т.к. информация о годе и сезоне Олимпийских игр содержится в полях Year и Season соответственно.  
2. Заменим для удобства значения NaN в поле Medal на строку "No". Пропуски данных в полях Age, Height, Weight для простоты заменим на медианные значения по соответствующим колонкам таблицы.  
3. Для сокращения потребления памяти уменьшим размерность целочисленных полей. Кроме того, поля Season и Medal сделаем категориальным, т.к. переменные в этох колонках могут принимать малое количество различных значений (2 и 4 соответственно).

```python
# 1. Удаление полей
df = df.drop(columns=['Name', 'Games'])
df.head()
```

```python
# 2. Обработка пропусков
df['Medal'] = df['Medal'].fillna('No')
for col_name in ('Age', 'Height', 'Weight'):
    df[col_name] = df[col_name].fillna(df[col_name].median())

df.head()
```

```python
# 3. Измнение типов
for col_name in ('Season', 'Medal'):
    df[col_name] = df[col_name].astype('category')

for col_name in ('ID', 'Age', 'Height', 'Weight', 'Year'):
    df[col_name] = df[col_name].astype('int32')


df.info()
```

```python
df.isnull().any().any()
```

Все пропуски удалены. Экономия памяти благодаря преобразованию типов - более 12 Мбайт.

## 5. Корреляция признаков

```python
df.corr(method='pearson')
```

```python
df.corr(method='kendall')
```

```python
df.corr(method='spearman')
```

```python
fig, ax = plt.subplots(1, 3, sharex='col', sharey='row', figsize=(18, 7))
sns.heatmap(df.corr(method='pearson'), ax=ax[0], annot=True, fmt='.2f', cmap='magma')
sns.heatmap(df.corr(method='kendall'), ax=ax[1], annot=True, fmt='.2f', cmap='magma')
sns.heatmap(df.corr(method='spearman'), ax=ax[2], annot=True, fmt='.2f', cmap='magma')
fig.suptitle('Корреляционные матрицы, построенные различными методами')
ax[0].title.set_text('Pearson')
ax[1].title.set_text('Kendall')
ax[2].title.set_text('Spearman')
plt.show()
```

Высокую степень корреляции имеет только одна пара признаков - рост и вес.
